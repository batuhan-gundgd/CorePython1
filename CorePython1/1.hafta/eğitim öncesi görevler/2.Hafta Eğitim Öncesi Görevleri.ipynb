### Soru 3
Dairenin alanını hesaplayan bir Python kodu yazınız. pi sayısını 3.14 olarak alınız ve yarıçapını kullanıcıdan alınız. Sonucu ekrana bastırınız.

````python
# 3.Sorunun çözümünü buraya yazabilirsin.
def daire_alani_hesapla(yaricap):
    pi = 3.14
    alan = pi * (yaricap ** 2)
    return alan

yaricap = float(input("Dairenin yaricapini girin: "))
alan = daire_alani_hesapla(yaricap)
print("Dairenin alani:", alan)
````

### Soru 4
x ve y adında iki liste oluşturunuz. Bu listeleri birbiryle yer değiştirerek yazdırınız.

```python
x = [1,2,3,4,5]
y = [5,4,3,2,1]
``` 

```python	
Output:
x = [5, 4, 3, 2, 1]
y = [1, 2, 3, 4, 5]
```


````python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

print("Önceki x:", x)
print("Önceki y:", y)

x, y = y, x

print("Yeni x:", x)
print("Yeni y:", y)
````
