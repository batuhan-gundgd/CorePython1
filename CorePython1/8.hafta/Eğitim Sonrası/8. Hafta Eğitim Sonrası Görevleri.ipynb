# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import time
def vki_hesapla(kilo,boy):
    vki = kilo / (boy ** 2)
    return round(vki,2)

def deger_araligi(vki):
    if vki <= 18.5:
        return "Zayıf"
    elif vki <= 24.9:
        return "Normal Kilolu"
    elif vki <= 29.9:
        return "Fazla Kilolu"
    elif vki <= 39.9:
        return "Obez"
    else:
        return "İleri Obez"
    
def cikti(isim,yas,kilo,boy,vki,deger):
    kayit = time.ctime(time.time())
    t1 = f"İsim: {isim}\n"
    t2 = f"Yas : {yas}\n"
    t3 = f"Vücut Kitle İndeksi : {vki}\n"
    t4 = f"Deger Araligi : {deger}\n"
    t5 = f"Kayıt Tarihi : {kayit}\n"
    output = t1 + t2 + t3 + t4 + t5
    return output

def main():
    giris_sayisi = int(input("Kisi Sayisi : "))
    for i in range(giris_sayisi):
        print(f"Kullanıcı kaydı {i+1}")
    isim = input("İsim Soyisim : ")
    yas = int(input("Yas : "))
    kilo = float(input("Kilo : "))
    boy = float(input("Boy(m) : "))
    vki = vki_hesapla(kilo,boy)
    deger = deger_araligi(vki)
    output = cikti(isim,yas,kilo,boy,vki,deger)
    with open("kayitlar.txt","a" , encoding="utf-8") as f:
        f.write(output)
        f.write("\n")
        f.close()
        print("Kaydedildi")
    print("Veriler 'kayitlar.txt' dosyasına kaydedildi ")

main()
 


    
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import time

def menu():
    def urunler():
        secim = input("Lütfen ürünü seçiniz:\n[1] Çilek 10TL\n[2] Elma 8TL\n[3] Böğürtlen 15TL\n[4] Mango 20TL\nSeçiminiz: ")
        if secim == '1':
            return 10
        elif secim == '2':
            return 8
        elif secim == '3':
            return 15
        elif secim == '4':
            return 20
        else:
            return None

    def tek_urun_fiyat(secim, adet):
        fiyat = urunler()
        if fiyat is not None:
            return fiyat * adet
        return None

    son_fiyat = 0
    satın_alınan_urunler = []
    satın_alma_adeti = []

    while True:
        eklenecek_urun = input("Hangi meyveyi eklemek istersiniz? (1-4) : ")
        if eklenecek_urun == "0":
            print("Fiş kaydediliyor...")
            break
        ekleme_adeti = int(input("Kaç adet eklemek istersiniz? : "))
        fiyat = tek_urun_fiyat(eklenecek_urun, ekleme_adeti)
        if fiyat is None:
            print("Hatalı seçenek!!")
        else:
            print(f"Satın alma ücreti : {fiyat} TL. \nSatın almayı onaylamak için [y] \nİptal için [n] giriniz.")
            check = input("Satın almayı onaylıyor musunuz? ")
            if check.lower() == "y":
                print("Satın alma onaylandı")
                print("-" * 40)
                son_fiyat += fiyat
                satın_alınan_urunler.append(eklenecek_urun)
                satın_alma_adeti.append(ekleme_adeti)
            elif check.lower() == "n":
                print("Satın alma iptal edildi.")
                print("-" * 40)
            else:
                print("Geçersiz seçenek!")

    return son_fiyat, satın_alma_adeti, satın_alınan_urunler

def main():
    print("YetBar'a hoş geldiniz!\nMenüyü görüntülemek için [1], Smoothie yapmak için [2], Çıkış için [0] tuşuna basınız.")
    while True:
        program = int(input("Tercihiniz: "))
        if program < 0 or program > 2:
            print("Yanlış bir tercih yaptınız!")
        elif program == 0:
            print("Çıkış yapılıyor.")
            break
        elif program == 1:
            menu()
        elif program == 2:
            ücret, urunler, adet = menu()
            with open("kasa.txt", "a", encoding="utf-8") as f:
                f.write(f"Tarih ve Saat: {time.ctime(time.time())}\n")
                f.write(f"Satın Alınan Ürünler: {', '.join(urunler)}\n")
                f.write(f"Adetler: {', '.join(map(str, adet))}\n")
                f.write(f"Toplam Satın Alma Ücreti: {ücret} TL\n")
                f.write("-" * 40 + "\n")

            kontrol = input("Başka bir işlem yapmak ister misiniz? (evet/hayır): ")
            if kontrol.lower() == "evet":
                continue
            else:
                print("Teşekkürler, ürünleriniz hazır.")
                break

main()






```
