# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız
ürün = {}
kac_urun = int(input("Kaç adet ürün bilgisi eklemek istersiniz? : "))

for i in range(kac_urun):
    ad = input("Ürün Adını Giriniz : ")
    fiyat = input("Ürünün Fiyat Bilgisi Giriniz : ")
    # Anahtar-değer çiftini sözlüğe ekleme kısmı oldukca onemliydi!!
    ürün[ad] = fiyat

x = str(ürün) #write() argument must be str, not dict

with open("urunler.txt" , "a" , encoding="utf-8") as f:
    f.write(x) #write() argument must be str, not dict

print(x)


````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız
eski_kelime = input("Değiştirilmesini istediğiniz kelimeyi giriniz : ")
yeni_kelime = input("Eski kelimenin yerini alacak kelimeyi giriniz : ")

# Dosyayı okuma ve yazma modunda aç
with open("urunler.txt", "r+", encoding="utf-8") as f:
    # Dosya içeriğini oku
    icerik = f.read()
    
    # Eski kelimeyi yeni kelime ile değiştir
    if eski_kelime in icerik:
        icerik = icerik.replace(eski_kelime, yeni_kelime) # replace() metodu önemliydi
        
        f.seek(0) # Dosyanın başına gidip içeriği yeniden yazmak için bu metodu kullandık
        f.write(icerik)
        f.truncate()  # Yeni içeriğin eski içerikten uzun olması durumunda fazlalıkları temizlemek için kullandık!!
        print(f"{eski_kelime} kelimesi {yeni_kelime} ile değiştirildi.")
    else:
        print("Girdiğiniz kelime bulunamadı.")
print(icerik)

````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage
    
class Bus(Vehicle):
    def __init__(self, name, speed, mileage):
        super().__init__(name, speed, mileage)
    
    def bilgiler(self): # bu kısma self koymayı unutmustum!!   
        return f"Name: {self.name} Speed: {self.speed} Mileage: {self.mileage}"

school_bus = Bus("School Volvo", 180, 12)
print(school_bus.bilgiler())

````
