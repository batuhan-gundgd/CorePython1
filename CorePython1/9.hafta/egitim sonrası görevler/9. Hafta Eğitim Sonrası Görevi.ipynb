Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
from random import randint 
tv_bilgi = {
    "Ürün Tipi": "LED TV",
    "Renk (Üreticiye Göre)": "Antrasit Gri",
    "Üretici Garantisi": "[30, 32, 35]",
    "Ekran Tipi" : "LED Ekran",
    "Boyut" : "139 cm / 55 inç",
    "Görüntü kalitesi" : "4K UHD"
}
ses_ayarları = []


class Kumanda():
    def __init__(self,tv_durum = "kapalı",kanal_listesi = ["a","b","c","d"],
            ses_ayarları = {
            "ses_seviyesi": 10,
            "ekolayzer": "stereo"}):
        self.tv_durum = tv_durum
        self.tv_bilgi = tv_bilgi
        self.kanal_listesi = kanal_listesi
        self.ses_ayarları = ses_ayarları
    
    def tv_ac(self):
        if (self.tv_durum == "kapalı"):
            print("Tv aciliyor...")
            self.tv_durum == "acik"
        else:
            print("Tv acik!")

    def tv_kapat(self):
        if self.tv_durum == "acik":
            self.tv_durum == "kapali"
        else:
            print("Tv kapali!")
    
    def kanalları_görüntüle(self):
        for kanal in self.kanal_listesi :
            print(kanal)
    
    def kanal_ekle(self):
        kanal = input("Eklemek istediğiniz kanalı yazın: ")
        self.kanal_listesi.append(kanal)
        print(f"{kanal} kanalı eklendi.")

    def tv_bilgileri():
        print("TV Bilgileri:")
        for key, value in tv_bilgi.items():
            print(f"{key}: {value}")
    
    def kanal_degistir(self):
        if self.kanal_listesi:
            index = randint(0, len(self.kanal_listesi) - 1)
            print(f"Şu anki kanal: {self.kanal_listesi[index]}")
        else:
            print("Kanal bulunmamaktadır.")
        
    def ses_seviyesi_artir(self):
        if self.ses_ayarları["ses_seviyesi"] <= 100:
            self.ses_ayarları["ses_seviyesi"] += 1
            print(f"Ses seviyesi artırıldı: {self.ses_ayarları['ses_seviyesi']}")
        else:
            print("Ses seviyesi zaten maximumda.")

    def ses_seviyesi_azalt(self):
        if self.ses_ayarları["ses_seviyesi"] > 0:
            self.ses_ayarları["ses_seviyesi"] -= 1
            print(f"Ses seviyesi azaltıldı: {self.ses_ayarları['ses_seviyesi']}")
        else:
            print("Ses seviyesi zaten minimumda.")

    def ses_modunu_degistir(self):
        if self.ses_ayarları["ekolayzer"] == "studio":
            self.ses_ayarları["ekolayzer"] = "vocal"
            print("ekolayzer : vocal")
        else:
            self.ses_ayarları["ekolayzer"] = "studio"
            print("ekolayzer : studio")

    @staticmethod # metoda dikkat!!
    def ses_ayarlarini_goster():
        print("**Ses ayarları**")
        # Burada 'ses_ayarları' sözlüğünü doğrudan kullanmalısınız.
        # 'self' kullanamazsınız çünkü bu metot statik bir metottur.
        print(f"Ses Seviyesi: {kumanda.ses_ayarları['ses_seviyesi']}")
        print(f"Ekolayzer: {kumanda.ses_ayarları['ekolayzer']}")



"""def menu(self):
        if self.otomat_durum == "kapalı":
              print("Once otomatı aciniz")
        else:
             for item in list(self.kanal_listesi()):
                print(f"{item.upper()} : {kanal_listesi} TL")"""

kumanda = Kumanda()

while True:
    islem = input("işlemi seçiniz:")
    if (islem == "q"):
        print("islem sonlandı!!!")
        break
    elif (islem == "1"):
        kumanda.tv_ac()
    elif (islem == "2"):
        kumanda.tv_kapat()
    elif (islem == "3"):
        kumanda.kanalları_görüntüle()
    elif (islem == "4"):
        kumanda.kanal_ekle()
    elif (islem == "5"):
        kumanda.kanal_degistir()
    elif (islem == "6"):
        Kumanda.tv_bilgileri() # 'kumanda' nesnesini degil 'Kumanda' sınıfını kullandık!!!
    elif (islem == "7"):
        kumanda.ses_seviyesi_artir()
    elif islem == "8":
        kumanda.ses_seviyesi_azalt()
    elif islem == "9":
        kumanda.ses_modunu_degistir()
    elif islem == "10":
        Kumanda.ses_ayarlarini_goster() #'kumanda' nesnesini degil 'Kumanda' sınıfını kullandık!!!
    else:
        print("Hatalı deger!!!.")




```
